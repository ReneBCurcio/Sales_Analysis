```python
import pandas as pd
```

```python
df = pd.read_csv("dataset_asimov.csv")
```

```python
df
```

```python
df.info()
```

```python
for i in range (0, len(df["Valor Pago"])-1):
    df["Valor Pago"] = df["Valor Pago"][i].split(" ")[1]
```

```python
df["Valor Pago"] = df["Valor Pago"].astype(int)
```

```python
for i in range (0, len(df["Duração da chamada"])):
    minutes, seconds = df["Duração da chamada"][i].split(":")
    result = int(minutes) + float(seconds)/60
    df["Duração da chamada"][i] = result
```

```python
df["Duração da chamada"] = df["Duração da chamada"].astype(float)
```
